```python
#!pip install Pillow==8.2.0
#!pip install fpdf==1.7.2
#!pip install tk==0.1.0


from PIL import ImageGrab
from fpdf import FPDF
from tkinter import Tk
from tkinter.filedialog import asksaveasfilename

import time
import os
import tempfile
import hashlib
import threading
import copy
```

```python
def getSaveFileNameDialog():
    Tk().withdraw()
    filename = asksaveasfilename(initialfile="capture", defaultextension=".pdf")
    return filename
```

```python
keyPressed = [False]
def keyListener():
    input()
    keyPressed[0] = True

def add_picture_to_pdf(pdf, im):
    pdf.add_page()
    fd, path = tempfile.mkstemp(suffix=".png")
    os.close(fd)
    im.save(path)
    pdf.image(path, 0,0,0,0, "PNG")
    os.unlink(path)

def main():
    old_hash = 0
    im_hash = 0
    
    pdf = None
    
    threading.Thread(target=keyListener).start()
    print("Press Enter to stop and save")
    
    while True:
                
        try:
            im = ImageGrab.grabclipboard()
        except OSError:
            print("Cannot load clipboard. Retrying...")
            im = None
            im_hash = 0

        if im != None:
            try:
                im_hash = hashlib.md5(im.tobytes()).hexdigest()
            except AttributeError:
                im = None
                im_hash = 0
                continue
            if(im_hash != old_hash):
                
                if pdf is None:
                    pdf = FPDF('L', 'pt', (im.size[1], im.size[0]))
                add_picture_to_pdf(pdf, im)
                print("new Image added to queue", im)
                old_hash = im_hash
        if(keyPressed[0]):
            break
        else:
            time.sleep(1)

    pdf.output(getSaveFileNameDialog(), 'F')
```

```python
main()
```
